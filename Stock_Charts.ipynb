```python
#this imports the yfinance library and renames it as yf for easier use in the code.
import yfinance as yf

#import the pyplot module from the matplotlib library, which is a plotting library for the Python programming language. It allows us to create a variety of charts and graphs, including line charts, scatter plots, bar charts, histograms, and more.
import matplotlib.pyplot as plt
```

```python
stock = 'ARGGY'
start = '2019-04-16' 
end = '2024-02-24'
data = yf.download(stock, start, end)
data
```

```python
# Plot the stock data
#data.Close.plot() - this plots the closing prices of the stock over time. The Close column of the data DataFrame contains the closing prices, and calling the plot() method on this column generates the plot.
#plt.show() - this displays the plot on the screen. plt is an alias for the matplotlib.pyplot library, and calling the show() function displays the plot.
#plt.show() - this displays the plot on the screen. plt is an alias for the matplotlib.pyplot library, and calling the show() function displays the plot.

data['Close'].plot()
plt.title(f"{ticker} Aston Martin Stock Price")
plt.xlabel("Date")
plt.ylabel("Price")
plt.savefig('Aston_Martin_stocks.pdf')
plt.show()
```
