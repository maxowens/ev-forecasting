```python
auto_stocks_dict = {'Tesla':'TSLA',
                   'Toyota':'TM',
                   'Ferrari':'RACE',
                    'Honda':'HMC',
                    'Ford': 'F',
                    'General Motors': 'GM',
                    'NIO': 'NIO',
                    'Polaris': 'PII',
                    'Kandi Technologies Group': 'KNDI',
                    'Electra Meccanica': 'SOLO',
                    'Arcimoto': 'FUV',
                   }

tickers = [val for val in auto_stocks_dict.values()]
```

```python
automotive_df = stock_data[stock_data['Symbol'].isin(tickers)]
automotive_df
```

```python
nasdaq_dict = {'Tesla': 'tsla_df',
               'Toyota': 'tm_df',
               'Ferrari': 'race_df',
              'Honda': 'hmc_df',
              'Ford': 'f_df',
              'General Motors': 'gm_df',
              'Kandi Technologies Group': 'kndi_df'}
```

```python

#Finding min & max dates for index
min_date = min(pd.to_datetime(globals()[df_name]['Date']).min() for df_name in nasdaq_dict.values())
max_date = max(pd.to_datetime(globals()[df_name]['Date']).max() for df_name in nasdaq_dict.values())

#Setting index to date range
date_range = pd.date_range(start=min_date, end=max_date, freq='D')
merged_df = pd.DataFrame(index=date_range)

# Merging dataframes
for company, df_name in nasdaq_dict.items():
    # Access the DataFrame using the variable name
    df = globals()[df_name]
    # Convert the 'Date' column to datetime if it's not already
    df['Date'] = pd.to_datetime(df['Date'])
    # Reset index to avoid issues with merging
    df = df.set_index('Date')
    # Merge the DataFrame onto the new DataFrame
    merged_df = pd.merge(merged_df, df['Close'], left_index=True, right_index=True, how='left')
    # Rename the 'Close' column to the company name
    merged_df.rename(columns={'Close': company}, inplace=True)


merged_df = merged_df.round(2)

merged_df
```
